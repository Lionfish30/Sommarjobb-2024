```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
temp = np.array([45.2, 97.8, 149.2,198.6])
current = np.array([1.01, 3.29, 5.590,7.883])

c = np.polyfit(current,temp,1)
x = np.linspace(0,8,1000)

plt.scatter(current,temp)
plt.plot(x,c[0]*x+c[1])
plt.ylabel("Temp")
plt.xlabel("Currnet")
plt.show()
print(c)
```

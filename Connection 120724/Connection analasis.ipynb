```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
def dataimporter_con(filename):
    data = pd.read_csv(filename,dtype=str)
    data["V"] = data["V1"]+"."+data["V2"]
    data["I"] = data["I1"]+"."+data["I2"]
    V = data["V"].astype(float).to_numpy()
    I = data["I"].astype(float).to_numpy()
    return V,I
```

```python
V1e,I1e = dataimporter_con("Connection 120724/Epoxy Chip 1.csv")
V2e,I2e = dataimporter_con("Connection 120724/Epoxy CHip 2.csv")
V3e,I3e = dataimporter_con("Connection 120724/Epoxy Chip 3 3'.csv")
V4e,I4e = dataimporter_con("Connection 120724/Epoxy Chip 4 3.csv")

V1o,I1o = dataimporter_con("Connection 120724/Over chip 1.csv")
V2o,I2o = dataimporter_con("Connection 120724/Over Cip 2.csv")
V3o,I3o = dataimporter_con("Connection 120724/Over Chip 3.csv")
V4o,I4o = dataimporter_con("Connection 120724/Over Chip 4.csv")



# print(data["Smu1.I[1][1]"].str.replace("E","e"))

# V = np.arange(-1,1.02,0.02)
# plt.plot(V,epoxychip4)
```

```python
plt.plot(V4e,I4e,label="Epoxy",color= "orange")
plt.plot(V3o,I3o,label="Over", color = "blue")
plt.plot(V4e,np.linspace(I4e[0],I4e[-1],101),linestyle = "dashed",color = "orange")
plt.plot(V3o,np.linspace(I3o[0],I3o[-1],101),linestyle = "dashed",color = "blue")

plt.legend()
plt.title("Connection IV measurement")
plt.xlabel("Volt(V)")
plt.ylabel("Current(I)")
plt.show()
```

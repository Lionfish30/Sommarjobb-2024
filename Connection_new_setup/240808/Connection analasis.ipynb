```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
def dataimporter_con(filename):
    data = pd.read_csv(filename,dtype=str)
    data["V"] = data["V1"]+"."+data["V2"]
    data["I"] = data["I1"]+"."+data["I2"]
    V = data["V"].astype(float).to_numpy()
    I = data["I"].astype(float).to_numpy()
    return V,I
```

```python
V2e,I2e = dataimporter_con("epoxy2.csv")
V3e,I3e = dataimporter_con("epoxy 3.csv")
Ve,Ie = dataimporter_con("epoxy clean.csv")

V1o,I1o = dataimporter_con("over 1.csv")
V2o,I2o = dataimporter_con("over 2.csv")
V3o,I3o = dataimporter_con("over deaD.csv")
V4o,I4o = dataimporter_con("over 4.csv")

VG,IG = dataimporter_con("graph pure.csv")


V1g,I1g = dataimporter_con("graph 1.csv")
V2g,I2g = dataimporter_con("graph 2.csv")
V3g,I3g = dataimporter_con("graph 3.csv")
V4g,I4g = dataimporter_con("graph 4.csv")

# print(data["Smu1.I[1][1]"].str.replace("E","e"))

# V = np.arange(-1,1.02,0.02)
# plt.plot(V,epoxychip4)
```

```python
plt.plot(Ve,Ie*10e5,label="Chip E",color = "blue")
# plt.plot(Ve,np.linspace(Ie[0]*10e5,Ie[-1]*10e5,101),linestyle = "dashed",color = "red")

plt.legend()
plt.title("Epoxy Connection")
plt.xlabel("Volt(V)")
plt.ylabel(r"Current($\mu$A)")
plt.show()
```
